### Tic Tac Toe

```python
import random
```

```python
class Board:
    
    def __init__(self,player_1,player_2):
        self.player_X= player_1
        self.player_O= player_2
        self.board= [' '] * 9 #(self.board= [' ',' ',' ',' ',' ',' ',' ',' ',' '] or best as written below        
                                #self.board= [' '*9] #Gives one item with 9 spaces)
        self.player_X_turn= random.choice([True, False])
    
    
    
    def display_board(self):
        b=self.board
        
        output= f"""
                
                {b[0]} | {b[1]} | {b[2]} 
                 ---------
                {b[3]} | {b[4]} | {b[5]} 
                 ---------
                {b[6]} | {b[7]} | {b[8]}
        
 
                 """
        print(output)
    
    def check_full(self):
                #  spaces= [x for x in self.board if x== ' ']
                #  return len(spaces)== 0 (Alternative code written below)
        count=0
        for char in self.board:
            if char==' ':
                count+=1
                
        if count== 0:
            return True
        else:
            return False
        
    def check_winner(self,char):
        b=self.board
        conditions= [
            [0,1,2],
            [3,4,5],
            [6,7,8],
            [0,3,6],
            [1,4,7],
            [2,5,8],
            [0,4,8],
            [2,4,6]
        ]
    
        for c in conditions:
            if char== b[c[0]]== b[c[1]]== b[c[2]]:
                return True
        return False
    
    def play(self):
        
        self.board= [' ']*9
        
        self.player_X.new_game()
        self.player_O.new_game()
        
        while True:
            
            if self.player_X_turn:
                player=self.player_X
                other_player=self.player_O
            else:
                player=self.player_O
                other_player=self.player_X
            
            char=player.char
            
            if player.type== 'human':
                print(f"Player {player.name}'s turn")
                self.display_board()
            
            move= player.make_move(self.board)
            
            while abs(move)>8 or self.board[move] != ' ':
                print("Invalid move. Choose another spot")
                self.display_board()
                move= player.make_move()
                
            self.board[move]= char
            
            if self.check_full():
                player.reward(5,self.board)
                other_player.reward(5,self.board)
                
                if player.type=='human':
                    print("Game Over. It's a draw")
                    break
                
            if self.check_winner(char):
                player.win()
                player.reward(10,self.board)
                other_player.reward(-10,self.board)
                break
            
            player.reward(-1,self.board)                
            self.player_X_turn= not self.player_X_turn
```

```python
class Player:
    
    def __init__(self,name,char):
        self.name=name
        self.char=char
        self.type= 'human'
        
    def reward(self,reward,board):
        print(f"{self.name} rewarded {reward}")
        
    def win(self):
        print(f"{self.name} won the game!")
                                      
    def make_move(self,board):
        move=input(f"Enter place to put the sign: {self.char}")
        return int(move)
    
    def new_game(self):
        print(f"{self.name} is {self.char}")
```

```python
class QAI(Player):
    
    def __init__(self,char, alpha=0.5,gamma=0.5,epsilon=0.1):
        self.alpha= alpha
        self.gamma= gamma
        self.epsilon= epsilon
        self.type= 'AI'
        self.name= 'AI'
        self.char=char
        self.q_table= {}
        
    def win(self):
        pass
    
    def new_game(self):
        last_board= (" ")*9
        last_move= None
        
    def get_Q(self,state,action):
        if self.q_table.get((tuple(state),action)) != None:
            return self.q_table[(state,action)]
        else:
            self.q_table((tuple(state),action)) == 5
            return self.q_table((state,action))
        
    def available_move(self,board):
        return [x for x in range(9) if board[x]== ' ']
        
    def make_move(self, board):

        self.last_board= board
        
        available_moves= self.available_move(board)
        
        rand= random.random()
        
        if rand < self.epsilon:
            self.last_move= random.choice(available_moves)
            return self.last_move
        
        q_values= [self.get_Q[tuple(board), action] for action in available_moves]
        max_q= max(q_values)        
        max_index= q_values.index(max_q)
                
        action_selected= available_moves[max_index]        
        self.last_move= action_selected
        
        return action_selected
    
    def reward(self,reward,board):
        board=tuple(board)
        last_board= self.last_board
        last_action= self.last_move
        if last_action is not None:
            prev= self.get_Q(last_board, last_action)
            available_moves= self.available_move(last_board)
            
            q_values= [self.get_Q[tuple(board), action] for action in available_moves]
            max_q= max(q_values)        
            
            self.q_table[(board, last_action)] = self.prev + self.alpha(self.gamma * reward + \
                                                                 max_q - prev)        
```

```python
player_1= QAI('X')
player_2= QAI('O')
```
